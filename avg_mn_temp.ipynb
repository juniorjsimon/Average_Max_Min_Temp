```python
#dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load csv
mn_temp_df = pd.read_csv("Resources/avg_temp_mn.csv")
mn_temp_df.head()
```

```python
#ensure date column is date format
#mn_temp_df['Date'] = pd.to_datetime(mn_temp_df['Date'])
```

```python
#start date and end date to filter in
start_date = '2019-12-01'
end_date = '2019-12-22'

#set a mask to apply the df filter
mask = (mn_temp_df['Date'] >= start_date) & (mn_temp_df['Date'] <= end_date)
```

```python
# Asign mask to df
mn_temp_df = mn_temp_df.loc[mask]
mn_temp_df.head()
```

```python
# Set x axis for maximum temperature
x_axis = np.arange(len(mn_temp_df))

#set tick locations
tick_locations = [value+0.5 for value in x_axis]

# set y axis for maximum temperature
y_axis_mx = mn_temp_df['Maximum Temperature degrees (F)']

# Set x and y axis for minimum temperature
x_axis = np.arange(len(mn_temp_df))
y_axis_mn = mn_temp_df['Minimum Temperature degrees (F)']

# set figure size and plot graphs
plt.figure(figsize=(9,9))
plt.plot(x_axis, y_axis_mx, label="Max Temp.(F)")
plt.plot(x_axis, y_axis_mn, label="Min Temp.(F)")

#plot tick location
plt.xticks(tick_locations, mn_temp_df["Date"], rotation="vertical")

# Draw a horizontal line with 0.25 transparency
plt.hlines(20, 0, 22, alpha=0.25, color='black')

# Adds a legend and sets its location to the lower right
plt.legend(loc="best")

```

```python
# Create a labels for the chart
plt.title("Max and Min Avg temp. from Dec 01-19 - Dec 22-19 in MN")
plt.ylabel("Fahrenheit")
plt.show()
```
